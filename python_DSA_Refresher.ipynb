Question 1: Number of seconds in a given number of days


```python
# Input number of days
days = int(input("Enter the number of days: "))

# Calculate seconds
# Number of days * 24 hours in a day * 60 minutes in an hour * 60 seconds in a minute
seconds = days * 24 * 60 * 60

# Results
print(f"There are {seconds} seconds in {days} days.")
```

Question 2: Calculate the volume of a sphere

```python
# Input radius
radius = float(input("Enter the radius of the sphere: "))

# Calculate volume V = (4/3)πr^3
# set 3.14159 as  π
pi = 3.14159
volume = (4/3) * pi * radius**3

# result
print(f"The volume of the sphere is {volume}")
```

Question 3: Compute the area and perimeter of a square using functions

```python
# Function to calculate the area of a square
def calculate_area(side):
    return side * side

# Function to Calculates the perimeter of a square
def calculate_perimeter(side):
    return 4 * side

# Input side length
side_length = float(input("Enter the side length of the square: "))

# Calculate area and perimeter
area = calculate_area(side_length)
perimeter = calculate_perimeter(side_length)

# results
print(f"Area of the square: {area}")
print(f"Perimeter of the square: {perimeter}")
```

Question 4: Determine whether a character is uppercase or lowercase

```python
# Funtion to check if a character is uppercase, lowercase, or not a letter
def check_case(character):
    if character.isupper():
        return "uppercase"

    elif character.islower():
        return "lowercase"

    # If the character is neither
    else:
        return "not a letter"

# Input character
char = input("Enter a character: ")

# To ensure only one character is entered at a time
if len(char) == 1:
    # Call the function to check the case
    result = check_case(char)
    print(f"The character '{char}' is {result}.")
else:
    print("Please enter only one character.")
```

Question 5: Python equivalent of the psuedocode

```python
# Initialize variables
x = 0
y = 20

# Repeat the loop until y is less than 6
while y >= 6:
    # Subtract 4 from y
    y = y - 4
    # Add 2/y to x
    x = x + 2/y

# Display final x
print(f"x = {x}")
```

Question 6: Input values into an array and calculate the average

```python
# Initialize an empty array
values = []

# Loop to input 5 values
for i in range(5):
    number = float(input(f"Enter value {i + 1}: "))
    values.append(number)

# Calculate the average
total = sum(values)
average = total / len(values)

# results
print(f"The average of the values is {average}")
```
